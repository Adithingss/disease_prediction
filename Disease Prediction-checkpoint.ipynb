```python
encoder = LabelEncoder()
data["prognosis"] = encoder.fit_transform(data["prognosis"])

#This will standardize all the String values of Prognosis into numerical values 
```

```python
X = data.iloc[:,:-1]   
y = data.iloc[:, -1]

#[:,:] This means all rows and all cols
#[:,1:] This means all rows and cols except the first col
#[:,:-1] This means all rows and cols except the last col
#[:, -1] This means all rows of col -1


#Alternative for the above code would be

# X = data.drop(columns='prognosis', axis=1)
# y = data['prognosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify= y, random_state=30)


#Here suppose there are equal number of values in every disease. Now it can happen that
#while splitting into train and test all the values can be transferred to train so there is 
#no data left for test. so we mentionstratify
```

```python
X_train
```

```python
models = [SVC(kernel='linear'), GaussianNB(), RandomForestClassifier(random_state=18)]
```

```python
def cross_Validation_model():
    
    for model in models:
        
        scores = cross_val_score(model, X, y, cv=10)
        
        print(model)
        print(f"Scores: {scores}")
        print(f"Mean Score: {np.mean(scores)}")
        print("="*30)
        
        
cross_Validation_model()
```

From the above output it can be seen that our machine learning algorithms
are performing ver well and the mean scores after k fold cross-validation
are also very high. 

```python
def training_testing_split():
    
    for model in models:
    
        #Training the model
        model.fit(X_train, y_train)
        
        #Evaluating the model
        test_data_prediction = model.predict(X_test)
        train_data_prediction = model.predict(X_train)
    
        accuracy_of_train_data = accuracy_score(y_train, train_data_prediction)
        accuracy_of_test_data = accuracy_score(y_test, test_data_prediction)
    
        print(f"Accuracy on Train data by {model}: {accuracy_of_train_data*100}")
        print(f"Accuracy on Test data by {model}: {accuracy_of_test_data*100}")
        print("\n")
        
        cf_matrix = confusion_matrix(y_test, test_data_prediction)
        plt.figure(figsize=(10,5))
        sns.heatmap(cf_matrix, annot=True)
        print(f"Confusion matrix for {model} on Test data")
        plt.show()
        
        print("=="*30)
    

training_testing_split()
```

```python
#Training the model on whole train data 

final_svm_model = SVC()
final_nb_model = GaussianNB()
final_rf_model = RandomForestClassifier(random_state = 18)

final_svm_model.fit(X, y)
final_nb_model.fit(X, y)
final_rf_model.fit(X, y)
```

```python
# Reading the test data

test_data = pd.read_csv("C:/Users/adity/Testing.csv")
```

```python
test_X = test_data.iloc[:,:-1]
test_Y = encoder.transform(test_data.iloc[:, -1])
```

```python
svm_preds = final_svm_model.predict(test_X)
nb_preds = final_nb_model.predict(test_X)
rf_preds = final_rf_model.predict(test_X)

```

```python
# mode is defined for numbers, and does not do any sort of object method deligation
# sum([np.array([1,2,3,4,5]), np.array([1,2,3,4,5])]) = array([ 2,  4,  6,  8, 10])
#The sum will give an output but it not the same while we use mode fun
# mode([np.array([1,2,3,4,5]), np.array([1,2,3,4,5])])
#The above code gives an error and it also shows the same error while using list

#The scipy mode makes a 2d array from your inputs, and iterates across the columns,
#determining the mode one column at a time. 
#I suspect [mode(args) for args in zip([arr1,arr2])] would do just as well.


final_preds = []
for i, j, k in zip(svm_preds, nb_preds, rf_preds):
    
    mode_value = mode([i, j, k])[0][0]
    final_preds.append(mode_value)


# The above code is simple version of below code to understand better    
# final_preds = [mode([i,j,k])[0][0] for i,j,k in zip(svm_preds, nb_preds, rf_preds)]
# final_preds
```

```python
print(f"Accuracy on Test Dataset by the combined model: {accuracy_score(test_Y, final_preds)*100}")

cf_matrix = confusion_matrix(test_Y, final_preds)
plt.figure(figsize = (10, 5))
sns.heatmap(cf_matrix, annot= True)

plt.title("Confusion Matrix on Combined model on Test Dataset")
plt.show()
```

```python
symptons = X.columns.values
#symptons
#print(list(enumerate(symptons)))
```

```python
symptom_index = {}

for index, value in enumerate(symptons):
    
    symptom_words = value.split("_")   #This will split the _ among the words 
    symptom = " ".join(symptom_words)   #The splitted words simply cannot be 
    #added to the dict. we join them using the join func 
    
    symptom_index[symptom] = index
    
 
#print(symptom_index)
```

```python
data_dict = {
    "symptom_index":symptom_index,  #This will give symptom with its index
    "predictions_classes":encoder.classes_  #"predictions_classes" is the array 
                                            #of unique classes that the encoder
                                            #was fitted on.
}

print(data_dict)
```

```python
def predictDisease(symptoms):
    
    symptoms = symptoms.split(",")  #This will split the input string into a list of symptoms
    
    input_data = [0]*len(data_dict['symptom_index'])
    for symptom in symptoms:
        index = data_dict['symptom_index'][symptom]
        input_data[index] = 1
    
    input_data = np.array(input_data).reshape(1,-1)
    
    rf_prediction = data_dict["predictions_classes"][final_rf_model.predict(input_data)[0]]
    nb_prediction = data_dict["predictions_classes"][final_nb_model.predict(input_data)[0]]
    svm_prediction = data_dict["predictions_classes"][final_svm_model.predict(input_data)[0]]
    
    final_prediction = mode([rf_prediction, nb_prediction, svm_prediction])[0][0]
    
    predictions = {
        "rf_model_prediction": rf_prediction,
        "nb_model_prediction": nb_prediction,
        "svm_model_prediction": svm_prediction,
        "final_prediction": final_prediction
    }
    
    return predictions
    
```

```python
#The reason for this reshaping is that the input format expected by the
# machine learning models is a 2D array, where each row represents an instance
# (in this case, a single instance of symptom inputs), and each column 
# represents a feature (in this case, a symptom).

# The resulting array has a shape of (1, n), where n is the number of symptoms in
# the symptom index dictionary. This numpy array is then used as input to the
# machine learning models for making predictions
```
